```python
#Dataset for this code is taken from UCI repositories
#URL: https://archive.ics.uci.edu/dataset/235/individual+household+electric+power+consumption

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import requests
import zipfile
import io


# Reading the data
df = pd.read_csv("/content/household_power_consumption.txt",
                 sep=';',
                 parse_dates={'datetime': ['Date', 'Time']},
                 dayfirst=True,
                 na_values='?',
                 low_memory=False)

df.head()

df = df.set_index('datetime')
df = df.dropna() # to remove the missing values

# Converting columns to numeric data
numeric_columns = ['Global_active_power', 'Global_reactive_power', 'Voltage',
                  'Global_intensity', 'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']
df[numeric_columns] = df[numeric_columns].astype(float)
df.head()
```

```python
plt.figure(figsize=(15, 10))

# 1. Daily average power consumption
daily_avg = df['Global_active_power'].resample('D').mean()
plt.subplot(2, 2, 1)
daily_avg.plot(title='Daily Average Global Active Power (kW)')
plt.xlabel('Date')
plt.ylabel('Power (kW)')

# 2. Hourly average power consumption
hourly_avg = df['Global_active_power'].groupby(df.index.hour).mean()
plt.subplot(2, 2, 2)
hourly_avg.plot(title='Hourly Average Global Active Power (kW)')
plt.xlabel('Hour of Day')
plt.ylabel('Power (kW)')

# 3. Sub-metering comparison
plt.subplot(2, 2, 3)
for col in ['Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']:
    df[col].resample('D').mean().plot(label=col)
plt.title('Daily Average Sub-metering')
plt.xlabel('Date')
plt.ylabel('Energy (watt-hour)')
plt.legend()

# 4. Voltage distribution
plt.subplot(2, 2, 4)
sns.histplot(df['Voltage'], bins=50)
plt.title('Voltage Distribution')
plt.xlabel('Voltage (V)')

plt.tight_layout()
plt.show()

#Correlation matrix
plt.figure(figsize=(10, 8))
correlation_matrix = df[numeric_columns].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix of Power Measurements')
plt.show()

print("\nDataset Basic Statistics:")
print(df.describe())
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt

df = pd.read_csv("/content/household_power_consumption.txt",
                 sep=';',
                 parse_dates={'datetime': ['Date', 'Time']},
                 dayfirst=True,
                 na_values='?',
                 low_memory=False)

# Handle NaN values
df = df.set_index('datetime')
df = df.interpolate(method='linear')
df = df.dropna()

# Resample to hourly
df_hourly = df.resample('H').mean()

# Add time-based features
df_hourly['hour'] = df_hourly.index.hour
df_hourly['day_of_week'] = df_hourly.index.dayofweek

print("NaN check after preprocessing:")
print(df_hourly.isnull().sum())

features = ['Global_active_power', 'Voltage', 'Global_intensity', 'hour', 'day_of_week']

median_power = df_hourly['Global_active_power'].median()
df_hourly['target_class'] = (df_hourly['Global_active_power'] > median_power).astype(int)

# Function to create sequences
def create_sequences(data, target, seq_length):
    X, y = [], []
    attending = False
    for i in range(len(data) - seq_length):
        seq = data[i:(i + seq_length), :]
        if np.any(np.isnan(seq)):
            continue
        X.append(seq)
        y.append(target[i + seq_length])
    return np.array(X), np.array(y)

# Prepare data
seq_length = 24
data = df_hourly[features].values
target = df_hourly['target_class'].values

# Scale features
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data)

# Create sequences
X, y = create_sequences(data_scaled, target, seq_length)

# Verify data
print(f"X shape: {X.shape}, y shape: {y.shape}")
print(f"NaNs in X: {np.any(np.isnan(X))}, NaNs in y: {np.any(np.isnan(y))}")

# Split data (80% train, 20% test)
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

model = Sequential([
    LSTM(100, activation='relu', input_shape=(seq_length, len(features)), return_sequences=True),
    Dropout(0.2),
    LSTM(100, activation='relu', return_sequences=True),
    Dropout(0.2),
    LSTM(50, activation='relu'),
    Dropout(0.2),
    Dense(64, activation='relu'),
    Dropout(0.2),
    Dense(1, activation='sigmoid')

])

# Compiling model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Early stopping to prevent overfitting
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Train model
history = model.fit(X_train, y_train,
                   epochs=50,
                   batch_size=32,
                   validation_split=0.1,
                   callbacks=[early_stopping],
                   verbose=1)

# Prediction
y_pred_prob = model.predict(X_test)
y_pred = (y_pred_prob > 0.5).astype(int).ravel()

# Calculating accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"LSTM Classification Performance:")
print(f"Accuracy: {accuracy:.4f}")

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Low', 'High'])
disp.plot(cmap='Blues')
plt.title('Confusion Matrix for LSTM Power Consumption Classification')
plt.show()

plt.figure(figsize=(12, 4))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('LSTM Training History (Loss)')
plt.xlabel('Epoch')
plt.ylabel('Binary Crossentropy Loss')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 4))
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('LSTM Training History (Accuracy)')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()

# Predicted vs actual values
plt.figure(figsize=(12, 6))
time_index = df_hourly.index[-len(y_test):]
plt.plot(time_index, y_test, label='Actual Class', color='blue', alpha=0.6)
plt.plot(time_index, y_pred, label='Predicted Class', color='red', alpha=0.6)
plt.title('LSTM: Predicted vs Actual Power Consumption Class (0=Low, 1=High)')
plt.xlabel('Time')
plt.ylabel('Class')
plt.legend()
plt.grid(True)
plt.show()

# Predicted vs actual values for 200hrs
plt.figure(figsize=(12, 6))
plt.plot(time_index[-200:], y_test[-200:], label='Actual Class', color='blue', alpha=0.6)
plt.plot(time_index[-200:], y_pred[-200:], label='Predicted Class', color='red', alpha=0.6)
plt.title('LSTM: Predicted vs Actual (Last 200 Hours)')
plt.xlabel('Time')
plt.ylabel('Class')
plt.legend()
plt.grid(True)
plt.show()
```
